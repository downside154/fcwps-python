```python
# Leap year: 4의배수 이면서, 100배수가 아니거나 400의 배수인 연도
# Constraint: 1999~2101
# output: 'leap_year' if i is leap_year else, i
year_list = [i for i in range(1999,2101+1)]
is_leap_list = []
# 2000 = leap_year
# 2100 = plain_year
for year in year_list:
    if year%4==0 and (year%100!=0 or year%400==0):
        is_leap_list.append('leap_year')
        #print('leap')
    else:
        is_leap_list.append(year)
        #print(year)
is_leap_list
```

```python
# Leap year with list comprehension
# Leap year: 4의배수 이면서, 100배수가 아니거나 400의 배수인 연도
# Constraint: 1999~2101
# output: 'leap_year' if i is leap_year else, i
years = [i for i in range(1999,2101+1)]
is_leap_list = [year for year in years]
# 2000 = leap_year
# 2100 = plain_year
is_leap_list
```

```python
# Leap year with list comprehension
# Leap year: 4의배수 이면서, 100배수가 아니거나 400의 배수인 연도
# Constraint: 1999~2101
# output: 'leap_year' if i is leap_year else, i
years = [i for i in range(1999,2101+1)]
is_leap_list = ['leap_year' if year%4==0 and (year%100!=0 or year%400==0)
                else year for year in years]
# 2000 = leap_year
# 2100 = plain_year
is_leap_list
```

```python
# Leap year with list comprehension
# Leap year: 4의배수 이면서, 100배수가 아니거나 400의 배수인 연도
# Constraint: 1999~2101
# output: 'leap_year' if i is leap_year else, i
years = [i for i in range(1999,2101+1)]

def is_leap(i):
    if i%4==0 and (i%100!=0 or i%400==0):
        return True
    return False

is_leap_list = ['leap_year' if is_leap(year)
                else year for year in years]
# 2000 = leap_year
# 2100 = plain_year
is_leap_list
```

```python
fruits = {'a':'apple', 'b':'banana', 'c':'coconut', 'd':'durian'}
over_six_fruits = {}
for k,v in fruits.items():
    if len(v) > 6:
        over_six_fruits['k'] = v
over_six_fruits
```

```python
list(fruits.items())
```

```python
# dict comprehension
fruits = {'a':'apple', 'b':'banana', 'c':'coconut', 'd':'durian'}
over_six_fruits = {k:v for k,v in fruits.items() if len(v)>6}
```

```python
over_six_fruits
```

```python
number_dict = {'pi':3.1415, 'e':2.71828, 'year':2021, 'month':2, 'day':1}
new_number_dict = {}
for k,v in number_dict.items():
    if v < 3:
        new_number_dict[k] = v ** 2
    else:
        new_number_dict[k] = v - 1
new_number_dict
```

```python
number_dict = {'pi':3.1415, 'e':2.71828, 'year':2021, 'month':2, 'day':1}
new_number_dict = {k:v**2 if v < 3 else v-1 for k,v in number_dict.items()}
new_number_dict
```
